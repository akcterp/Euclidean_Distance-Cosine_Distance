```python
import numpy as np
import pandas as pd
```

```python
#Normalize functions
# 2 different types of normalizations 
def l1_normalize(v):
    norm = np.sum(np.abs(v))
    return v / norm

def l2_normalize(v):
    norm = np.sqrt(np.sum(np.square(v)))
    return v / norm
```

```python
# For calculating the Euclidean Distance

def euclidean_distance(x,y,normalize=False):
    if normalize==True:
        x=l2_normalize(x)
        y=l2_normalize(y)
    return np.sqrt(np.sum((x-y)**2))
```

```python
# For calculating the Cosine Similarity

def cosine_similarity(x, y):
    return np.dot(x, y) / (np.sqrt(np.dot(x, x)) * np.sqrt(np.dot(y, y)))
```

```python
TwentyNames = ['Lionel_Messi','Cristiano_Ronaldo','Neymar','Luis_Suárez','Eden_Hazard',
               'LeBron_James','Michael_Jordan','Kevin_Durant','Kobe_Bryant','Stephen_Curry',
               'Prasanta_Chandra_Mahalanobis','C._R._Rao','Raj_Chandra_Bose','Jayanta_Kumar_Ghosh','Arup_Bose',
               'C._V._Raman','A._P._J._Abdul_Kalam','Vikram_Sarabhai','Srinivasa_Ramanujan','Satyendra_Nath_Bose'
              ]
```

```python
article = []

for i in TwentyNames:
    article.append(wikipedia.page(i).content)
```

```python
#importing Count vectorizer to create a token array of counts of words(frequency) in each text document
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
#creating count vectorizers and accessing all the pages and getting frequencies
cv=CountVectorizer()
Cv_array = cv.fit_transform(article).toarray()
```

```python
#creating a celebrity list
people_euc = pd.DataFrame(TwentyNames,columns = ['Famous People'])
#people_euc
```

```python
people_euc_norm =  people_euc.copy()
people_cosine = people_euc.copy()
```

```python
# both functions take 2 names from the TwentyNames, returns the euclidian dist and cosine similarity
def dist1(a,b,normalize=False):
    x = Cv_array[TwentyNames.index(a)]
    y = Cv_array[TwentyNames.index(b)]
    return euclidean_distance(x,y,normalize)

def dist2(a,b):
    x = Cv_array[TwentyNames.index(a)]
    y = Cv_array[TwentyNames.index(b)]
    return cosine_similarity(x,y)
```

```python
# without normalization
for i in TwentyNames:
    people_euc[i] = people_euc['Famous People'].apply(lambda x : dist1(x,i))
    
# with normalization
for i in TwentyNames:
    people_euc_norm[i] = people_euc_norm['Famous People'].apply(lambda x : dist1(x,i,True))
    
# cosine similarity 
for i in TwentyNames:
    people_cosine[i] = people_cosine['Famous People'].apply(lambda x : dist2(x,i))

```

```python
people_euc.set_index('Famous People', inplace = True) 
people_euc_norm.set_index('Famous People', inplace = True) 
people_cosine.set_index('Famous People', inplace = True) 
```

```python
def knn_euclidean(Col,k=3,normalize=False):
    indx= []
    col1 = []
    if (normalize==True):
        indx = people_euc_norm.loc[people_euc_norm.index != Col,Col]
    else:
        indx = people_euc.loc[people_euc.index != Col,Col]
    for i in indx.index:
        col1.append([indx[i],i])   
    col1.sort()
    vals = []
    for i in range(k):
        vals.append(col1[i][1])
    return vals
```

```python
def knn_cosine(Col,k=3):
    col1 = []
    indx = people_cosine.loc[people_cosine.index != Col,Col]
    for i in indx.index:
        col1.append([indx[i],i])   
    col1.sort(reverse=True)  
    vals = []
    for i in range(k):
        vals.append(col1[i][1])   
    return vals
```

```python
three_nearest = pd.DataFrame(TwentyNames,columns = ['Famous'])
three_nearest.head()
```

```python
three_nearest['3NN_euclidean_withiout normalization'] = three_nearest['Famous'].apply(lambda x : knn_euclidean(x,3))
three_nearest['3NN_euclidean_with normalization'] = three_nearest['Famous'].apply(lambda x : knn_euclidean(x,3,True))
three_nearest['3NN_cosine similarity'] = three_nearest['Famous'].apply(lambda x : knn_cosine(x,3))
```
